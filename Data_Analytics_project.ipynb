# Data-Cleaning

```python
Evergreen = pd.read_csv('supermarket_sales - Data_Project.csv')
Evergreen
```

```python
Evergreen = pd.DataFrame(Evergreen)
Evergreen
```

Steps to be followed in order to further the dataset more interprative and make necessary tranformation:

Step-1) Checking if there are any NULL values within the dataset.

```python
NUll_Values = Evergreen.isnull()
NUll_Values
# There are no null values in the database.
```

Step-2) Getting a detailed stastical description of each column that has quantitative data undern the dataset.


```python
Evergreen.describe()
```

Step-3) Making corrections to the date column by changing the format of certain dates.

```python
Evergreen['Date'] = pd.to_datetime(Evergreen['Date'])
Evergreen['Date']
```

Step-4) Rounding up float values upto 2 decimal places.

```python
Coulmns_list = ['Tax 5%','Total','Date','Time','Payment','cogs','gross margin percentage','gross income']
Evergreen[Coulmns_list] = Evergreen[Coulmns_list].round(2)
Evergreen[Coulmns_list]
```

Step-5) Applying Binary-text modeling on columns for customer type,and payment mechanism.

```python
###a)customer type:
######1)Member-column:
Evergreen['Member'] = Evergreen['Customer type'].str.contains('Member',case = False)
Evergreen['Member'] = Evergreen['Member'].replace(True,1,regex=True)
Evergreen['Member'] = Evergreen['Member'].replace(False,0,regex=True)

######2)Normal-column:
Evergreen['Normal'] = Evergreen['Customer type'].str.contains('Normal',case = False)
Evergreen['Normal'] = Evergreen['Normal'].replace(True,1,regex=True)
Evergreen['Normal'] = Evergreen['Normal'].replace(False,0,regex=True)

###b) Creating a  columns for payment mechanism.
########1) Cash:
Evergreen['Cash'] = Evergreen['Payment'].str.contains('Cash',case = False)
Evergreen['Cash'] = Evergreen['Cash'].replace(True,1,regex=True)
Evergreen['Cash'] = Evergreen['Cash'].replace(False,0,regex=True)

#######2)Ewallet:
Evergreen['Ewallet'] = Evergreen['Payment'].str.contains('Ewallet',case = False)
Evergreen['Ewallet'] = Evergreen['Ewallet'].replace(True,1,regex=True)
Evergreen['Ewallet'] = Evergreen['Ewallet'].replace(False,0,regex=True)

#######3)Credit card:
Evergreen['Credit card'] = Evergreen['Payment'].str.contains('Credit card',case = False)
Evergreen['Credit card'] = Evergreen['Credit card'].replace(True,1,regex=True)
Evergreen['Credit card'] = Evergreen['Credit card'].replace(False,0,regex=True)

Evergreen
```

Maximum stock calculation:

So the minimum monthly demand is 31 units.So the reoder point should be around 40 units in order to avoid out of stock situation. So it will help the store managers of all the branches in ordering the managing the stock in the warehouse. 

For this reason, we will universerly accepted maximum inventory level formula:-

Reorder point: The point when your inventory needs to be replenished so you donâ€™t run into a stockout situation.
Minimum consumption: The smallest amount of demand you expect during a time frame.
Minimum order lead time: The minimum number of days it will take to receive your replenishment at your warehouse once the PO is placed.

Model information:
Reorder point = 40 units
minimum consumption = 31 units
